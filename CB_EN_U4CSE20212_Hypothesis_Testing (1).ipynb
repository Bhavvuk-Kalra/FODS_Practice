```python
#Q1
import numpy as np
import scipy.stats as st

mu = 100
Sd = 15
n = 30
xbar = 140
alpha = 0.05

#Ho mu = 100
#H1 mu !=100

z = ((xbar - mu)*(n)**0.5)/Sd
print(z)

z_critical = st.norm.ppf(alpha/2)
print(z_critical)

if(z>abs(z_critical) or z<z_critical):
  print("Reject")
else:
  print("Fail to Reject")
```

```python
#Q1 - p value
import numpy as np
import scipy.stats as st

mu = 100
Sd = 15
n = 30
xbar = 140
alpha = 0.05

#Ho mu = 100
#H1 mu !=100

z = ((xbar - mu)*(n)**0.5)/Sd
print(z)

p_val = 2*(1-st.norm.cdf(abs(z)))
print(p_val)

if(p_val<alpha):
  print("Reject")
else:
  print("Fail to Reject")
```
