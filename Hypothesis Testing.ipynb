== Z test ==

```python
import numpy as np
import scipy.stats as st
import pandas as pd

mean_population = 46
SD_population = 4.2
n = 30
mean_sample = 48.6
alpha = 0.05

z_critical = abs(st.norm.ppf(alpha))
z = (mean_sample-mean_population)/(SD_population/np.sqrt(n))

if(z>=z_critical):
  print("Reject Null Hypothesis!!")
else:
  print("Failed to Reject Null Hypothesis")

a = pd.DataFrame({"Component":["Z","Z Critical"],"Value":[z,z_critical]})
a
```

== Z test using P-Value ==

```python
mean_population = 8
n = 32
mean_sample = 8.2
SD_population = 0.6
alpha = 0.05

z = (mean_sample-mean_population)/(SD_population/np.sqrt(n))
p = (1-st.norm.cdf(abs(z)))*2

if(p>alpha):
  print("Failed to Reject Null Hypothese")
else:
  print("Reject Null Hypothesis!!")
```
