# Binomial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.binomial(12, 0.2, 20000)
```

```python
sum(np.random.binomial(12, 0.2, 20000)==4)/20000
```

```python
sum(np.random.binomial(12, 0.2, 20000)==4)/20000
```

```python
a = pd.Series(np.random.binomial(12, 0.2, 20000)).value_counts().index
a
b = pd.Series(np.random.binomial(12, 0.2, 20000)).value_counts().values
b
plt.bar(a,b)
```

# Uniform

```python
np.random.uniform(0,20,20000)
```

```python
a = np.random.uniform(0,20,20000)
b = (sum(a>10)/20000)
c = (sum(a<10)/20000)
print(b,c,b+c)
```

```python
a = np.random.uniform(0,20,20000)
b = (sum(a>5)/20000)
c = (sum(a<10)/20000)
print(b,c,b-c)
```

# Normal

```python
mean = 10
std = 2

a = np.random.normal(mean,std,20000)
a
```

```python
b = np.exp((-(((a-mean)/std)**2)/2)/std*(2*np.pi)**0.5)
b
```

```python
plt.bar(a,b)
```
