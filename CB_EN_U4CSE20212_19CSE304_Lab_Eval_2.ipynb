```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
#Q1
#H0 : Number of purple plants = 3/4*total_number_plants (75%)
#H1 : Number of purple plants != 3/4*total_number_plants

# therefore is a 2 tail test...
```

```python
total_number_plants = 929
total_number_plants
```

```python
number_sims = 2000 - 212
number_sims
```

```python
number_of_purple = 705 + 12
number_of_purple
```

```python
arr = []
np.random.multinomial(total_number_plants,[0.75,0.25])

for i in range(number_sims):
  arr.append(np.random.multinomial(total_number_plants,[0.75,0.25]))

a = pd.DataFrame(arr)
a
```

```python
a["Total"] = a[0]+a[1]
a["Avg"] = a[0]/a["Total"]
a
```

```python
#Q2
#we can use z test as the population size >30
len(a)
```

```python
#Q3
#Observed Statistics
obs_statistic = (number_of_purple/total_number_plants)
obs_statistic
```

```python
#Q4
averages = a["Avg"]
averages
```

```python
print(averages.mean())
print(averages.std())
```

```python
z = (obs_statistic-(averages.mean()))*(len(averages)**0.5)/(averages.std())
z
```

```python
alpha = 0.5

z_critical = st.norm.ppf(alpha/2)
print(z_critical)

if(z<z_critical or z>abs(z_critical)):
  print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
```

# Part B

```python
#Using the critical Value method, population SD is known, so Z test
```

```python
sd = 13.46
mu = 60
n = 30
xbar = 65.10
alpha = 0.05

#H0: mu = 60
#H1: mu != 60

z = ((xbar - mu)*(n)**0.5)/sd
print(z)

z_critical = st.norm.ppf(alpha/2)
print(z_critical)

if(z>abs(z_critical) or z<z_critical):
  print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
```
